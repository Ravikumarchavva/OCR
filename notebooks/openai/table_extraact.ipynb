```python
import pdfplumber

def extraxt_tables_from_pdf(file_path):
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                for row in table:
                    print(row)

extraxt_tables_from_pdf(file_path)
```

```python
# get images from pdf plumber and display them
def extract_images_from_pdf(file_path):
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            print(page.images)
extract_images_from_pdf(file_path)
```

```python
import pandas as pd

def extract_data_from_excel(file_path):
    df = pd.read_excel(file_path, engine="xlrd")  # Load the sheet
    df = df.dropna(how="all")  # Drop empty rows
    return df.to_dict(orient="records")  # Convert to structured format

file_path = ROOT_DIR / 'data' / 'mark' / 'Fw__Examples_of_QC_reports_for_setting_up_templates_in_Leadtools' / 'DPS9 PO 171876  ISS PO 212979.xls'
extract_data_from_excel(file_path)
```
